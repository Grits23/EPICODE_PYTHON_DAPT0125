```python
#L'azienda Object SpA ha creato una lista di quanti oggetti ha venduto ogni mese nell'ultimo anno:

lst = [2000, 5500, 7200, 4320, 1280, 1900, 2500, 3900, 6410, 8150, 7100, 5350]

#trasformiamola in un array NumPy (casting): lst = np.array(lst) e rispondiamo alle domande del CEO della Object SpA:
# • qual è stata la vendita massima mensile? E quella minima? • quali sono
# le vendite mensili maggiori di 5000 oggetti? E quante ne sono? • quali sono
# le vendite minori di 3000 oggetti? 5 • in media quanti oggetti sono stati venduti al mese

lst = np.array(lst)
print("Vendita massima mensile =", lst.max())
print("Vendita minima mensile =", lst.min())
print("Vendita maggiore di 5000 oggetti =", lst[lst > 5000])
print("Vendita minore di 3000 oggetti =", lst[lst < 3000])
print("Media vendite mensili =", lst.mean())
```

```python
#Consideriamo il seguente dizionario:
fatturati_dict = {1997: 12000, 1998: 15000, 1999: 20000, 2000: 23000, 2001: 25000, 2002: 17000, 2003: 14000, 2004: 21000}
#Consideriamo ora la seguente Series:
fatturati_series = pd.Series([12000, 15000, 20000, 23000, 25000, 17000, 14000, 21000], index=range(1997, 2005))
#Possiamo accedere alle stesse informazioni nello stesso modo:
fatturati_dict[1997]
fatturati_series[1997]
#Dunque qual è la differenza tra i due tipi di dato? 6 Cosa potremmo fare con la Series che non possiamo fare con il dizionario?


# Accesso ai dati
print("Accesso al valore 1997 nel dizionario:", fatturati_dict[1997])
print("Accesso al valore 1997 nella Series:", fatturati_series[1997])

# Operazione su tutta la serie (es: aumento del 10%)
print("\nFatturati aumentati del 10%:")
print(fatturati_series * 1.1)

# Filtro logico: fatturati > 20000
print("\nAnni con fatturato > 20000:")
print(fatturati_series[fatturati_series > 20000])

# Statistiche descrittive
print("\nMedia fatturati:", fatturati_series.mean())
print("Somma fatturati:", fatturati_series.sum())
print("Deviazione standard:", fatturati_series.std())

# Grafico dell’andamento dei fatturati
print("\nVisualizzazione del grafico")
fatturati_series.plot(title="Andamento Fatturati 1997–2004")

```

```python
#Scrivere un programma che, data una lista di numeri, fornisca in output i tre numeri più grandi;
#gestire il caso in cui la lista sia più corta di tre, e quando uno o più dei numeri selezionati sono uguali.

import random

numeri_random = random.choices(range(1, 10), k=random.randint(1, 9)) #creo una lista di valori e lunghezza raandomizzati
lista=[]

for numero in numeri_random: #riempio la lista
  lista.append(numero)
print("La lista è:", lista)

lista = pd.Series(lista) #trasformo la lista in formato series
max_output = lista
max_output = max_output.sort_values(ascending=False).head(3)

if len(lista) < 3:
  print("La lista è più corta di tre numeri.")
elif max_output.nunique() < 3:
  print("Uno o più numeri selezionati sono uguali:", max_output.nunique())
else:
  print("I tre numeri più grandi sono:")
  print(max_output)
```

```python
#Scrivere un programma che • in input acquisisce una lista di numeri e un numero K
#• in output, dovrà restituire la media di tutti i numeri nella lista maggiori o uguali a K
#• se non ce ne dovesse essere nessuno, dovrà stampare a schermo un messaggio adeguato.

input_lenght = int(input("Inserisci la lunghezza della lista: "))
list_values = random.choices(range(1, 100), k=input_lenght)
lista=[]
K_value = int(input("Inserisci il valore di K: "))

for numero in list_values:
  lista.append(numero)
print()
print("La lista è:", lista)
print("Il valore di K è:", K_value)

output = np.array(lista)
output_high = output[output[:] >= K_value]
output_mean = output_high.mean()
if output_high.size == 0:
  print("Non ci sono numeri maggiori o uguali a K nella lista.")
else:
  print()
  print("Numeri maggiori o uguali a K:", output_high)
  print("Media dei numeri maggiori o uguali a K:", output_mean)

```

```python
#Abbiamo una lista con i guadagni degli ultimi 12 mesi (supponiamo da Gennaio a Dicembre):
#guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]
#dobbiamo confrontare, stampando tutto a video, il guadagno di ogni mese con la media dei guadagni precedenti,
#e specificare nell'output se il guadagno attuale è maggiore o minore della media dei precedenti.
#Esempio di un possibile output: Mese 1: 100 € Mese 2: 90 € (media prec: 100 € - il guadagno attuale è minore)
#Mese 3: 70 € (media prec: 95 € - il guadagno attuale è minore)


guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]
mesi = ["Gennaio", "Febbraio", "Marzo", "Aprile", "Maggio", "Giugno",
        "Luglio", "Agosto", "Settembre", "Ottobre", "Novembre", "Dicembre"]

dataframe = pd.DataFrame({'Mese': mesi, 'Guadagno': guadagni})
media_precedenti = []

print("\n Confronto Guadagni Mensili:\n")

for i in range(len(guadagni)):
    if i == 0:
        media_precedenti.append(0)
        print(f"{mesi[i]:<10}: {guadagni[i]:>4} € | (nessuna media precedente)")
    else:
        media = sum(guadagni[:i]) / i
        media_precedenti.append(media)
        confronto = "maggiore" if guadagni[i] > media else "minore"
        print(f"{mesi[i]:<10}: {guadagni[i]:>4} € | media prec.: {media:.2f} € - guadagno {confronto}")

# Aggiungiamo la colonna al DataFrame
dataframe['Media Guadagni Precedenti'] = [round(m, 2) for m in media_precedenti]

# Visualizzazione del DataFrame finale
print("\n Tabella riepilogativa:\n")
print(dataframe.to_string(index=False))
```
